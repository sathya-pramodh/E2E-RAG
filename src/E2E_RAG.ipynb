```python
from gensim.models import KeyedVectors
from word_vectors import convert, FileType

convert("./wiki-news-300d-1M.vec", "./wiki-news-300d-1M.w2v", FileType.W2V, FileType.FASTTEXT)
convert("./wiki-news-300d-1M-subword.vec", "./wiki-news-300d-1M-subword.w2v", FileType.W2V, FileType.FASTTEXT)
convert("./crawl-300d-2M-subword.vec", "./crawl-300d-2M-subword.w2v", FileType.W2V, FileType.FASTTEXT)
convert("./crawl-300d-2M.vec", "./crawl-300d-2M.w2v", FileType.W2V, FileType.FASTTEXT)
```

```python
!ls | grep .w2v
```

```python
w2v_model1 = KeyedVectors.load_word2vec_format("./wiki-news-300d-1M.w2v", binary=True)
w2v_model1.save('./wiki-news-300d-1M.bin')
w2v_model2 = KeyedVectors.load_word2vec_format("./wiki-news-300d-1M-subword.w2v", binary=True)
w2v_model2.save('./wiki-news-300d-1M-subword.bin')
w2v_model3 = KeyedVectors.load_word2vec_format("./crawl-300d-2M.w2v", binary=True)
w2v_model3.save('./crawl-300d-2M.bin')
w2v_model4 = KeyedVectors.load_word2vec_format("./crawl-300d-2M-subword.w2v", binary=True)
w2v_model4.save('./crawl-300d-2M-subword.bin')
```

```python
query_word = "Athena"
num_neighbours = 5

print(w2v_model1.most_similar(positive=[query_word], topn=num_neighbours))
print(w2v_model2.most_similar(positive=[query_word], topn=num_neighbours))
print(w2v_model3.most_similar(positive=[query_word], topn=num_neighbours))
print(w2v_model4.most_similar(positive=[query_word], topn=num_neighbours))
```
